```python
import keras
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.normal(0,0.55,size=(1000,))

y = 0.1 * x + 0.3 + np.random.normal(0,0.03,size=(1000,))

plt.figure(figsize=(15,5))
plt.scatter(x,y,alpha=0.3)
plt.show()
```

```python
model = keras.Sequential(name='LinearR')

model.add(keras.layers.Dense(1,input_dim=1))

model.compile(optimizer='sgd',loss=['mse'],metrics=['mse'])
```

```python
model.fit(x, y, batch_size=128, epochs=100, validation_data=(x,y), verbose=1)
```

```python
model.summary()
```

```python
model.get_weights(),model.weights
```

```python
a,b = model.get_weights()

y_pred = x * a.flatten() + b

plt.figure(figsize=(15,5))
plt.scatter(x,y, alpha=0.3)
plt.plot(x, y_pred, 'r')
plt.show()
```
