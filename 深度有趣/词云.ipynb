```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import jieba
```

```python
text = open('data/wordcloud.txt',encoding='utf-8').read()
text = ' '.join(jieba.cut(text))
mask = np.asarray(Image.open('data/wordcloud.png'))
wc = WordCloud(font_path="C:/Windows/Fonts/simhei.ttf",
              mask=mask).generate(text)
```

```python
plt.imshow(wc)
plt.axis('off')
plt.show()
```

```python
wc.to_file("data/wordcloud.jpg")
```
