```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
# Create a black image
img = np.zeros((512,512,3), np.uint8)
# Draw a diagonal blue line with thickness of 5 px
plt.imshow(cv2.line(img,(0,0),(511,511),(255,0,0),5))
```

```python
plt.imshow(cv2.ellipse(img,(256,256),(100,50),0,0,180,255,-1))
```

```python
pts = np.array([[10,5],[20,30],[70,20],[50,10]], np.int32)
pts = pts.reshape((-1,1,2))
plt.imshow(cv2.polylines(img,[pts],True,(0,255,255)))
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
plt.imshow(cv2.putText(img,'OpenCV',(10,500), font, 4,(255,255,255),2,cv2.LINE_AA))
```

```python
import cv2
import numpy as np
cap = cv2.VideoCapture(0)
while(1):
    # Take each frame
    _, frame = cap.read()
    # Convert BGR to HSV
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    # define range of blue color in HSV
    lower_blue = np.array([110,50,50])
    upper_blue = np.array([130,255,255])
    # Threshold the HSV image to get only blue colors
    mask = cv2.inRange(hsv, lower_blue, upper_blue)
    # Bitwise-AND mask and original image
    res = cv2.bitwise_and(frame,frame, mask= mask)
    cv2.imshow('frame',frame)
    cv2.imshow('mask',mask)
    cv2.imshow('res',res)
    k = cv2.waitKey(5) & 0xFF
    if k == 27:
        break

cv2.destroyAllWindows()
cap.release()
```

```python
img1 = cv2.imread('meixi.jpg')
img2 = cv2.imread('opencv-logo.jpg')
plt.subplot(121),plt.imshow(img1,'gray'),plt.title('img1')
plt.subplot(122),plt.imshow(img2,'gray'),plt.title('img2')
```

```python
rows,cols,channels = img2.shape
roi = img1[0:rows, 0:cols]
plt.imshow(cv2.cvtColor(roi,cv2.COLOR_BGR2RGB))
```

```python
img2gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
plt.imshow(cv2.bitwise_not(img2gray),cmap='gray')
```

```python
ret, mask = cv2.threshold(img2gray, 190, 255, cv2.THRESH_BINARY)
ret, plt.imshow(mask,cmap='gray')
```

```python
mask_inv = cv2.bitwise_not(mask)
plt.imshow(mask_inv,cmap='gray')
```

```python
img1_bg = cv2.bitwise_and(roi,roi,mask = mask)
plt.imshow(img1_bg)
```

```python
img2_fg = cv2.bitwise_and(img2,img2,mask = mask_inv)
plt.imshow(img2_fg)
```
